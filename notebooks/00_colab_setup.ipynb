## ðŸ’¡ Colab Pro Tips

### Performance Optimization
- **Use GPU**: Always enable GPU for faster training
- **Batch size**: Start with smaller batches (16-32) to avoid OOM errors
- **Mixed precision**: Use `torch.cuda.amp` for faster training
- **Persistent workers**: Set `persistent_workers=True` in DataLoader

### Session Management
- **Save frequently**: Models and results to Google Drive
- **Runtime limits**: Free tier has 12-hour limit, Pro has 24-hour
- **Idle timeout**: Sessions timeout after 30-90 minutes of inactivity
- **Reconnect**: Use `!nvidia-smi` to check if GPU is still available

### Memory Management
- **Clear cache**: `torch.cuda.empty_cache()` to free GPU memory
- **Monitor usage**: Use `!nvidia-smi` to check GPU memory
- **Gradient accumulation**: Simulate larger batches with less memory

### Collaboration
- **Share notebooks**: Use Google Drive sharing or GitHub
- **Version control**: Commit changes to GitHub regularly
- **Data sharing**: Use shared Google Drive folder for datasets

## ðŸ”§ Troubleshooting

### Common Issues

**GPU not available**
- Solution: Runtime > Change runtime type > GPU

**Out of memory errors**
- Solution: Reduce batch size, use gradient accumulation
- Code: `torch.cuda.empty_cache()`

**Session disconnected**
- Solution: Re-run setup cells, models/data saved to Drive persist

**Import errors**
- Solution: Re-run pip install cell, check Python path

**Slow training**
